```python
from bs4 import BeautifulSoup
import requests
from requests import get
import pandas as pd
import numpy as np
import re
import dateutil.parser
```

```python
response = requests.get('https://www.the-numbers.com/box-office-star-records/domestic/lifetime-acting/top-grossing-leading-stars')
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
actors_list = []
top_100_actors = soup.find_all(id="page_filling_chart")[1].find_all("a")
for text in top_100_actors:
    print(text.text)
    actors_list.append(text.text)
```

```python
top_actors_list = pd.DataFrame(actors_list)
top_actors_list.drop(top_actors_list.tail(6).index, inplace=True)
top_actors_list.rename(columns={0:"actors"}, inplace=True)
```
